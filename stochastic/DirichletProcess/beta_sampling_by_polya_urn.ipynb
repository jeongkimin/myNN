```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def polya_urn(n0, n1, T): #fixed
    history = []
    cur_n0 = n0
    cur_n1 = n1
    for t in range(T):
        p0 = cur_n0 / (cur_n0 + cur_n1)
        if p0 > random.uniform(0, 1):
            cur_n0 += 1
            history.append(1.0)
        else :
            cur_n1 += 1
            history.append(0.0)
    history = np.array(history)
    return history


p_mean = []
for i in range(200000):
    p_mean.append(polya_urn(alpha, beta, 2000).mean())
```

```python
def beta_pdf(x, _alpha, _beta):
    const = math.gamma(_alpha + _beta)/(math.gamma(_alpha) * math.gamma(_beta))
    return const * ((x)**(_alpha - 1.)) * ((1. - x)**(_beta - 1.))


plt.hist(p_mean, bins=100, density=True, rwidth=1.0)
x = np.linspace(0, 1, 100)
plt.plot(x, beta_pdf(x, alpha, beta), label='beta('+str(alpha)+','+str(beta)+')')
plt.legend()
plt.show()
```
